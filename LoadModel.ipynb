```python
# load the model from disk
import pickle

filename = './drive/MyDrive/delesmell/delesmellmodel.sav'
delmodel = pickle.load(open(filename, 'rb'))
```

```python
df.drop('is_brain', inplace=True, axis=1)
df
```

```python
df.drop('NAME', inplace=True, axis=1)
```

```python
predict_data = df.head(4)
```

```python
predict_data = predict_data.values.reshape((predict_data.shape[0],predict_data.shape[1],1))
```
